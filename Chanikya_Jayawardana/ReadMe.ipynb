## Project Overview

This project explores nonadiabatic molecular dynamics using the Libra package.

The objective was to understand the workflow for trajectory-based nonadiabatic dynamics and investigate how the electronic dynamics change with different electronic coupling strengths and different propagation methods.

The calculations are based on the two-level Holstein model included in the Libra tutorials.

---

## Calculations Performed

The following simulations were carried out:

1. FSSH with weak electronic coupling
2. FSSH with medium electronic coupling
3. FSSH with strong electronic coupling
4. Ehrenfest dynamics with medium coupling

For each simulation the model Hamiltonian was modified by changing the coupling parameter (V) while keeping the remaining model parameters unchanged.

---

## Workflow

The workflow for each calculation is:

1. Select the model Hamiltonian.
2. Define the Hamiltonian parameters.
3. Select the nonadiabatic dynamics method.
4. Define simulation parameters.
5. Specify the nuclear and electronic initial conditions.
6. Run the dynamics.
7. Analyze and plot the results.

## Results

Each notebook generates plots including:

- Potential, kinetic and total energies
- Nuclear coordinates
- Nuclear momenta
- Nuclear forces
- Phase-space trajectories
- Adiabatic electronic populations
- Diabatic electronic populations
- Surface hopping populations (FSSH)
- MASH populations

These plots were used to compare the behavior of different coupling strengths and propagation methods.

---

## Observations

Increasing the electronic coupling modifies the population transfer between the electronic states.

FSSH produces stochastic transitions (surface hopping) between adiabatic states, while Ehrenfest dynamics evolves the electronic wavefunction continuously on an averaged potential surface.

The simulations illustrate how electronic coupling strength influences the degree of nonadiabatic population transfer.
